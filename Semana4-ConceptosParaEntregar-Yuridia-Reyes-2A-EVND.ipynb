```python
#Comandos para sacar el promedio o MEDIA ARITMETICA:
#Cargamos el Dataset#

import pandas as pd
data = pd.read_csv('SoftwareProject_DataSet.csv')

#Creacion de una muestra aleatoria#

sample = data.sample(n=200, random_state=42)

#Verificacion del tamaño de la muestra#

print("La muestra tiene", len(sample), "observaciones")
sample.head()
```

```python
#Seleccion del valor que se pretende trabajar

code_lines_of_sample = sample["Líneas de código"]

#Creacion de la Funcion que realizara la operacion

def mean_calculator(data_list):
    sum_of_values = sum(data_list)
    sample_size = len(data_list)
    mean = sum_of_values / sample_size
    return mean

code_lines_mean = mean_calculator(code_lines_of_sample)

print("La media de los códigos de la muestra es:", code_lines_mean)
```

```python
#Comandos para sacar la MEDIANA:
#Creacion de la funcion que realizara la operacion

def median_calculator(data_list):
    sorted_list = sorted(data_list)
    sample_size = len(data_list)
    halfed_sample_size = sample_size // 2

    if sample_size % 2 == 0:
        return (sorted_list[halfed_sample_size-1] + sorted_list[halfed_sample_size]) / 2
    else:
        return sorted_list[halfed_sample_size]

code_line_median = median_calculator(code_lines_of_sample)

print(f"Mediana manual: {code_line_median} vs Pandas: {code_lines_of_sample.median()}")
```

```python
#Comandos para sacar la MEDIA RECORTADA:
def sample_cutter(sample, cut_percentage):
    sorted_sample = sorted(sample)
    sample_length = len(sample)
    index_to_cut = (cut_percentage / 100) * sample_length

    elements_to_cut = int(index_to_cut) if index_to_cut.is_integer() else int(index_to_cut) + 1

    sample_cutted = sorted_sample[elements_to_cut : sample_length - elements_to_cut]

    print("el tamaño de la muestra es: ", len(sample_cutted))

    return sample_cutted


cutted_mean = mean_calculator(sample_cutter(code_lines_of_sample, 10))

print("La Media recortada es: ", cutted_mean)
```
